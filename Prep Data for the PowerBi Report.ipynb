```python
def data(sheet_name):
    return pd.read_excel(xls, sheet_name)
```

```python
doc = data('Unique and Total Records')
doc = doc.loc[doc['Domain'] != 'Social History']
doc
```

```python
doc['doc_order'] = [1 if i=="Document" else 2 if i=='Allergy' else 3 if i=='Immunization' else
                    4 if i=='Medication' else 5 if i=='Problem' else 6 if i=='Procedure' else 
                    7 if i=='Result' else 8                       
                    for i in doc.Domain]
doc = doc.sort_values(by='doc_order')
doc
```

```python
alr = data('Allergy_SbCS')
alr
```

```python
alr['KPI'] = np.where(alr['KPI'].str.contains('2.16'), 'Present', alr['KPI'])
alr = alr.groupby('KPI', as_index=False).sum()
alr
```

```python
alr.columns
alr.rename(columns=({'KPI':'Observation'}), inplace=True)
alr.columns
```

```python
alr_sb_Text = alr[(alr.Sb_Text != 0)][['Observation', 'Sb_Text', '%_Sb_Text']].reset_index()
alr_sb_Text.drop('index', axis=1, inplace=True)
alr_sb_Text

alr_SC = alr[(alr.SC != 0)][['Observation', 'SC', '%_SC']].reset_index()
alr_SC.drop('index', axis=1, inplace=True)
alr_SC
alr_SCS = alr[(alr.SCS != 0)][['Observation', 'SCS', '%_SCS']].reset_index()
alr_SCS.drop('index', axis=1, inplace=True)
alr_SCS
alr_R_Text = alr[(alr.R_Text != 0)][['Observation', 'R_Text', '%_R_Text']].reset_index()
alr_R_Text.drop('index', axis=1, inplace=True)
alr_R_Text
alr_RC = alr[(alr.RC != 0)][['Observation', 'RC', '%_RC']].reset_index()
alr_RC.drop('index', axis=1, inplace=True)
alr_RC
alr_RCS = alr[(alr.RCS != 0)][['Observation', 'RCS', '%_RCS']].reset_index()
alr_RCS.drop('index', axis=1, inplace=True)
alr_RCS
alr_Sv_Text = alr[(alr.Sv_Text != 0)][['Observation', 'Sv_Text', '%_Sv_Text']].reset_index()
alr_Sv_Text.drop('index', axis=1, inplace=True)
alr_Sv_Text
alr_SVC = alr[(alr.SVC != 0)][['Observation', 'SVC', '%_SVC']].reset_index()
alr_SVC.drop('index', axis=1, inplace=True)
alr_SVC
alr_SVCS = alr[(alr.SVCS != 0)][['Observation', 'SVCS', '%_SVCS']].reset_index()
alr_SVCS.drop('index', axis=1, inplace=True)
alr_SVCS
```
